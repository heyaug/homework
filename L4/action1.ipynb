```python
import numpy as np
#转移矩阵
a = np.array([[0,0,0,1/3,0,0], 
            [1/4,0,0,0,1/2,0],
            [0,1,0,1/3,1/2,0],
            [1/4,0,0,0,0,1],
            [1/4,0,1,1/3,0,0],
            [1/4,0,0,0,0,0]])
```

```python
# 初始PR值
b = np.array([1/4,1/4,1/4,1/4,0,0])
w = b

# 简化模型
def work(a, w):
    for i in range(100):
        w = np.dot(a, w)
    return w
        
# 随机模型
def random_work(a, w, n):
    d = 0.85
    for i in range(100):
        w = (1-d)/n + d*np.dot(a, w)
    return w
```

```python
print("简化模型的PR值：",dict(zip(["A","B","C","D","E","F"],work(a, w))))
# 随机模型迭代100次的PR值
print("随机模型的PR值：",dict(zip(["A","B","C","D","E","F"],random_work(a, w,6))))
```

```python
# 使用networkX计算节点的pagerank
import networkx as nx
import matplotlib.pyplot as plt

# 创建有向图
G = nx.DiGraph()   
# 设置有向图的边集合
edges = [("A","B"),("A","D"),("A","E"),("A","F"),("B","C"),("C","E"),("D","A"),("D","C"),("D","E"),("E","B"),("E","C"),("F","D")]
# 在有向图G中添加边集合
for edge in edges:
    G.add_edge(edge[0], edge[1])

# 有向图可视化
layout = nx.spring_layout(G)
nx.draw(G, pos=layout, with_labels=True, hold=False)
plt.show()
```

```python
pr = nx.pagerank(G, alpha=1)
print("简化模型的PR值：", dict(sorted(pr.items(),key=lambda x:x[0])))
pr = nx.pagerank(G, alpha=0.8)
print("随机模型的PR值：", dict(sorted(pr.items(),key=lambda x:x[0])))
```

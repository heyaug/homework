#### 链家app 爬虫

```python
import base64
import hashlib
import requests
import time
import random

# headers里的Authorization,对url签名
def generateAuthorization(url):
    v0 = {i.split("=")[0]: i.split("=")[1] for i in url.split("?")[1].split("&")}
    v1 = sorted(v0.items(), key=lambda x: x[0], reverse=False)
    v2 = "93273ef46a0b880faf4466c48f74878f" + "".join([i[0] + "=" + i[1] for i in v1])
    v3 = hashlib.sha1(v2.encode('utf-8')).hexdigest()
    v4 = "20170324_android" + ":" + v3
    v5 = base64.b64encode(v4.encode("utf-8"))
    return v5.decode()

# 随机长度数字
def randomNum(Len):
    return "".join([random.choice("0123456789") for i in range(Len)])

# 随机长度字符串
def randomStr(Len):
    return "".join([random.choice("abcdefghijklmnopqrstuvwxyz0123456789") for i in range(Len)])

# 链家app楼盘数据接口，输入区域行政代码和页数
def price(cityid,page):
    ts=int(time.time())
    url = f"https://app.api.lianjia.com/newhouse/apisearch?has_filter=0&city_id={city_id}&from=search_result&from_container=0&page={page}&request_ts={ts}"
    headers = {'Page-Schema': 'newhouse%2Fprojectlist',
               'User-Agent': f'HomeLink9.12.0;{randomStr(7)} {randomStr(2)}; Android 5.1.1',
               'Lianjia-Channel': 'Android_Huawei',
               'Lianjia-Device-Id': '8650'+randomNum(12),
               'Device-Info': 'scale=1.2;screenwidth=720;screenheight=1280',
               'Lianjia-City-Id': city_id,
               'Lianjia-Version': '9.12.0',
               'Authorization': generateAuthorization(url),
               'Lianjia-Im-Version': '1.3.3',
               'Lianjia-NH-AB': '',
               'Host': 'app.api.lianjia.com',
               'Connection': 'Keep-Alive',
               'Accept-Encoding': 'gzip'}
    return requests.get(url, headers=headers, timeout=5).json()
```

```python
city_id="150200"
page=0
data=price(city_id,page)
# 区域楼盘总数
total_count=data["data"]["resblock_list"]["total_count"]
# 是否有下一页，停止条件
has_more_data=data["data"]["resblock_list"]["has_more_data"]
# 楼盘列表
house=data["data"]["resblock_list"]["list"]
```
